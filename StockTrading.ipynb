```python
#########################################  INDICATORS


def BOLBANDS(df): 
    upper, middle, lower = talib.BBANDS(df['Close'], matype=MA_Type.T3)
    return upper,middle,lower

def SMA(df):
    output = talib.SMA(df['Close'], timeperiod=100)
    return output

def MOM5(df):
    MOM5 = talib.MOM(df['Close'], timeperiod=5)
    return MOM5

def TSF14(df):
    real = TSF(df['Close'], timeperiod=14)
    return real

def calcADX(df):
    real = ADX(df['High'], df['Low'], df['Close'], timeperiod=14)
    return real

def TEMA(df):
    real = TEMA(df['Close'], timeperiod=30)
    return real
    
def MIDPRICES(df):
    real = MIDPRICE(df['High'], df['Low'], timeperiod=50)
    return real

def MIDPRICE1(df):
    real = MIDPRICE(df['High'], df['Low'], timeperiod=100)
    return real

def MIDPRICE3(df):
    real = MIDPRICE(df['High'], df['Low'], timeperiod=200)
    return real

def Simple_Moving_Averages():
    df["SMA1"] = df['Close']'].rolling(window=10).mean()
    df["SMA2"] = df['Close']'.rolling(window=50).mean()
    df["SMA3"] = df['Close'].rolling(window=100).mean()
```

```python
def Two_Crows(df):
    #The upside gap two crows is viewed by chartists as a somewhat ominous pattern, since it potentially signals that the index or security may be rolling over as its upward move ends and a downtrend begins.
    integer = CDL2CROWS(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Three_Black_Crows(df):
    #The bearish three black crows reversal pattern starts at or near the high of an uptrend, with three black bars posting lower lows that close near intrabar lows. This pattern predicts that the decline will continue to even lower lows, perhaps triggering a broader-scale downtrend. The most bearish version starts at a new high (point A on the chart) because it traps buyers entering momentum plays. According to Bulkowski, this pattern predicts lower prices with a 78% accuracy rate.4﻿
    integer = CDL3BLACKCROWS(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Three_Inside_Up_Down(df):    
    integer = CDL3INSIDE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Three_Line_Strike(df):
    #According to Bulkowski, this reversal predicts higher prices with an 84% accuracy rate
    integer = CDL3LINESTRIKE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Three_Outside_Up_Down(df):
    #Three outside up/down are patterns of three candlesticks that often signal a reversal in trend.
    
    integer = CDL3OUTSIDE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Three_Stars_In_The_South(df):
    
    
    integer = CDL3STARSINSOUTH(df['Open'], df['High'], df['Low'], df['Close'])

    
    #print(integer)
    return integer

def Three_Advancing_White_Soldiers(df):
    
    
    integer = CDL3WHITESOLDIERS(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Abandoned_Baby(df):
    #The bullish abandoned baby reversal pattern appears at the low of a downtrend, after a series of black candles print lower lows.8﻿ The market gaps lower on the next bar, but fresh sellers fail to appear, yielding a narrow range doji candlestick with opening and closing prints at the same price. A bullish gap on the third bar completes the pattern, which predicts that the recovery will continue to even higher highs, perhaps triggering a broader-scale uptrend. According to Bulkowski, this pattern predicts higher prices with a 70% accuracy rate.
    
    integer = CDLABANDONEDBABY(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer

def Advance_Block(df):
    
    
    integer = CDLADVANCEBLOCK(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Belt_hold(df):
   
    
    integer = CDLBELTHOLD(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Breakaway(df):
    
    
    integer = CDLBREAKAWAY(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Closing_Marubozu(df):
    
    
    integer = CDLCLOSINGMARUBOZU(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Concealing_Baby_Swallow(df):
    
    
    integer = CDLCONCEALBABYSWALL(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Counter_Attack(df):
    
    
    integer = CDLCOUNTERATTACK(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Dark_Cloud_Cover(df):
    
    
    integer = CDLDARKCLOUDCOVER(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer

def Doji(df):
    
    
    integer = CDLDOJI(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Doji_Star(df):
    
    
    integer = CDLDOJISTAR(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Dragonfly_Doji(df):
    
    integer = CDLDRAGONFLYDOJI(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Engulfing_Pattern(df):
    
    integer = CDLENGULFING(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer



def Evening_Doji_Star(df):
    
    
    integer = CDLEVENINGDOJISTAR(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer


def Evening_Star(df):
    #The bearish evening star reversal pattern starts with a tall white bar that carries an uptrend to a new high.6
    #The market gaps higher on the next bar, but fresh buyers fail to appear, yielding a narrow range candlestick. A gap down on the third bar completes the pattern, which predicts that the decline will continue to even lower lows, perhaps triggering a broader-scale downtrend. According to Bulkowski, this pattern predicts lower prices with a 72% accuracy rate.7﻿
    
    integer = CDLEVENINGSTAR(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer

def Up_Down_gap_side_by_side_white_lines(df):
    
    
    integer = CDLGAPSIDESIDEWHITE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Gravestone_Doji(df):
    
    
    integer = CDLGRAVESTONEDOJI(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer 




def Hammer(df):
    
    
    integer = CDLHAMMER(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer 


def Hanging_Man(df):
    
    
    integer = CDLHANGINGMAN(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Harami_Pattern(df):
    
    
    integer = CDLHARAMI(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Harami_Cross_Pattern(df):
    
    
    integer = CDLHARAMICROSS(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def High_Wave_Candle(df):
    
    
    integer = CDLHIGHWAVE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Hikkake_Pattern(df):
    
    
    integer = CDLHIKKAKE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def Modified_Hikkake_Pattern(df):
    
    integer = CDLHIKKAKEMOD(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Homing_Pigeon(df):
    
    integer = CDLHOMINGPIGEON(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer


def Identical_Three_Crows(df):
    
    integer = CDLIDENTICAL3CROWS(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer

def In_Neck_Pattern(df):

    integer = CDLINNECK(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Inverted_Hammer(df):
    
    

    integer = CDLINVERTEDHAMMER(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Kicking(df):   
     

    integer = CDLKICKING(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Kicking2(df):

    integer = CDLKICKINGBYLENGTH(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Ladder_Bottom(df):
    
    integer = CDLLADDERBOTTOM(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Long_Legged_Doji(df):

    integer = CDLLONGLEGGEDDOJI(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Long_Line_Candle(df):
    

    integer = CDLLONGLINE(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Marubozu(df):

    integer = CDLMARUBOZU(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Matching_Low(df): 

    integer = CDLMATCHINGLOW(df['Open'], df['High'], df['Low'], df['Close'])
    
    #print(integer)
    return integer
def Mat_Hold(df): 

    integer = CDLMATHOLD(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer
def Morning_Doji_Star(df):

    integer = CDLMORNINGDOJISTAR(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer

def Morning_Star(df):

    integer = CDLMORNINGSTAR(df['Open'], df['High'], df['Low'], df['Close'], penetration=0)
    
    #print(integer)
    return integer


def On_Neck_Pattern(df): 

    integer = CDLONNECK(df['Open'], df['High'], df['Low'], df['Close']) 
    #print(integer)
    return integer


def Piercing_Pattern(df): 

    integer = CDLPIERCING(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Rickshaw_Man(df):
 
    integer = CDLRICKSHAWMAN(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Rising_Falling_Three_Methods(df):

    integer = CDLRISEFALL3METHODS(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Separating_Lines(df):
 
    integer = CDLSEPARATINGLINES(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Shooting_Star(df):
 
    integer = CDLSHOOTINGSTAR(df['Open'], df['High'], df['Low'], df['Close'])

    #print(integer)
    return integer

def Short_Line_Candle(df):

    integer = CDLSHORTLINE(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Spinning_Top(df):
 
    integer = CDLSPINNINGTOP(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Stalled_Pattern(df):

    integer = CDLSTALLEDPATTERN(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Stick_Sandwich(df):

    integer = CDLSTICKSANDWICH(df['Open'], df['High'], df['Low'], df['Close'])    
    #print(integer)
    return integer


def Takuri(df):

    integer = CDLTAKURI(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer


def Tasuki_Gap(df):

    integer = CDLTASUKIGAP(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer
 

def Thrusting_Pattern(df):
    
    integer = CDLTHRUSTING(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer

def Tristar_Pattern(df):
 
    integer = CDLTRISTAR(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer

def Unique_3_River(df):
 
    integer = CDLUNIQUE3RIVER(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer

def Upside_Gap_Two_Crows(df):
 
    integer = CDLUPSIDEGAP2CROWS(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer

def Upside_Downside_Gap_Three_Methods(df):
     
    integer = CDLXSIDEGAP3METHODS(df['Open'], df['High'], df['Low'], df['Close'])
    #print(integer)
    return integer














#e bearish two black gapping continuation pattern appears after a notable top in an uptrend, with a gap down that yields two black bars posting lower lows.5﻿ This pattern predicts that the decline will continue to even lower lows, perhaps triggering a broader-scale downtrend. According to Bulkowski, this pattern predicts lower prices with a 68% accuracy rate.

#Two_Crows(df)
#Three_Black_Crows(df)
#Three_Inside_Up_Down(df)
#Three_Line_Strike(df)
#Three_Outside_Up_Down(df)
#Three_Stars_In_The_South(df)
#Three_Advancing_White_Soldiers(df)
#Abandoned_Baby(df)
#Advance_Block(df)
#Belt_hold(df)
#Breakaway(df)
#Closing_Marubozu(df)
#Concealing_Baby_Swallow(df)
#Counterattack(df)
#Dark_Cloud_Cover(df)
#Doji(df)
#Doji_Star(df)
#Dragonfly_Doji(df)
#Engulfing_Pattern(df)

```
